```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_file(path):
    data = []
    with open(path) as f:
        lines = f.readlines()
        lines.pop(0)
        for line in lines:
            split_line = line.split()
            data.append([int(split_line[0]), int(split_line[1])])
    return data
```

```python
def draw(kruskal, prim, density):
    fig = plt.figure(figsize=(10, 8))
    plt.plot([item[0] for item in kruskal], [item[1] for item in kruskal], label='Kruskal algorithm')
    plt.plot([item[0] for item in prim], [item[1] for item in prim], label='Prim algorithm')
    plt.title("Minimum Spanning Tree, graph density =" + str(density))
    plt.xlabel("Number of nodes")
    plt.ylabel("Time in microseconds")
    plt.legend()
    plt.savefig('./data/'+str(density)+'.png')
```

```python
for i in range(4, 10):
    kruskal = read_file('./data/kruskal_0_'+str(i)+'.txt')
    prim = read_file('./data/prim_0_'+str(i)+'.txt')
    draw(kruskal, prim, round(i*0.1, 2))
```
